```python
# Importing essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style for better aesthetics
sns.set(style="whitegrid")

# Load the dataset
df = pd.read_csv("creditcard.csv")

# Display the first few rows of the dataset
df.head()

```

```python
# Check the shape of the data
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")

# View basic information about each column, including data types
df.info()

# Display summary statistics for numerical features
df.describe()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values[missing_values > 0])
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values[missing_values > 0])
```

```python
# Plot the class distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df)
plt.title('Class Distribution')
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)')
plt.ylabel('Count')
plt.show()

# Display the proportion of each class
class_counts = df['Class'].value_counts(normalize=True)
print("Class Proportion:\n", class_counts)
```

```python
# Plot the distribution of 'Time'
plt.figure(figsize=(10, 5))
sns.histplot(df['Time'], bins=50, kde=True)
plt.title('Distribution of Transaction Time')
plt.xlabel('Time (seconds)')
plt.show()

# Plot the distribution of 'Amount'
plt.figure(figsize=(10, 5))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amount')
plt.xlabel('Amount ($)')
plt.show()
```

```python
# Boxplot for 'Amount' split by 'Class'
plt.figure(figsize=(8, 6))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Transaction Amount by Class')
plt.ylim(0, 500)  # Limiting y-axis for better visualization
plt.show()
```

```python
# Correlation matrix
corr_matrix = df.corr()

# Plot heatmap of the correlation matrix
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Boxplot for 'Amount' to detect outliers
plt.figure(figsize=(8, 6))
sns.boxplot(df['Amount'])
plt.title('Boxplot for Transaction Amount')
plt.show()
```

```python
# Boxplot for 'Time' to detect outliers
plt.figure(figsize=(8, 6))
sns.boxplot(df['Time'])
plt.title('Boxplot for Transaction Time')
plt.show()
```

```python
# Distribution of 'Amount' by class
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="Amount", hue="Class", bins=50, kde=True)
plt.title("Distribution of Transaction Amount by Class")
plt.xlabel("Amount ($)")
plt.show()

# Distribution of 'Time' by class
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="Time", hue="Class", bins=50, kde=True)
plt.title("Distribution of Transaction Time by Class")
plt.xlabel("Time (seconds)")
plt.show()
```
